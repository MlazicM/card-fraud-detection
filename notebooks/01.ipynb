```python
import pandas as pd

df = pd.read_csv("../data/raw/creditcard.csv")
df.head()
```

```python
df['Class'].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt

fraud = df[df["Class"] == 1]
legit = df[df["Class"] == 0]

plt.hist(legit["Amount"], bins=100, alpha=0.5, label="legit")
plt.hist(fraud["Amount"], bins=100, alpha=0.5, label="fraud")
plt.legend()
plt.show()
```
